```python
# Import the text file with error handling
try:
    with open('Harry Potter.txt', encoding='utf-8-sig', errors='replace') as text:
        text0 = text.read()
        print(text0[:10000])  # Print the first 10,000 characters
except UnicodeDecodeError as e:
    print(f"Error reading the file: {e}")
```

```python
# Delete all the URLs and Emails in the file
text00 = re.sub(r'www\S+', '', text0)
text01 = re.sub(r'http\S+', '', text00)
text02 = re.sub(r'\S+@\S+', '', text01)
print(text02[0:10000])
```

```python
# Delete all the number in the file
text1 = ''.join([i for i in text02 if not i.isdigit()]) # Call join on an empty string '' to concatenate without separators
print(text1[0:10000])
```

```python
# Delete all the Punctuation marks
text2 = text1.translate(str.maketrans('','',string.punctuation))
print(text2[0:10000])
```

```python
# Convert text to LOWERCASE
text3 = text2.lower()
print(text3[0:10000])
```

```python
# Seperate each works with a space and save the words to a list named "data"
data = text3.split()
print(data[0:100])
```

```python
# print out how many words you get after data cleaning
print(len(data))
```

```python
#Add Value 1 to each word in data list
data_pairs = []
for item in data:
  temp = item + ' ' + '1'
  data_pairs.append(temp)
print(data_pairs[0:100])
```

```python
# now we will group the sublists with same keys
# the format : data pairs = ['a 1', 'a 1', 'b 1', ....]
data_pairs.sort()
print(data_pairs[0:10000])
```

```python
# Removing duplicates to find all distinct words in the list
data_pairs_red = list(dict.fromkeys(data_pairs))
print(data_pairs_red)
print(len(data_pairs_red))
```
